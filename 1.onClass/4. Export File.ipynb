```python
//  Vấn đề: Khi tạo một file mới thì sẽ khá mất thời gian để import và export một quá trình cứ lặp đi lặp lại
// * Giải pháp: Sử dụng đọc file: Đọc hết file index.js, lặp qua nó rồi đẩy ra hết
// * Nghĩa là trong tương lai chúng ta không cần đi làm thủ công nữa
```

```python
// Đầu ra:
// const truncateTitle = require("./truncateTitle");
// const joinArray = require("./joinArray");
// module.exports = { truncateTitle, joinArray };

/**
 * Phương thức đọc thư mục:
 * - Dùng đọc đồng bộ thì mới export ra thẳng được, chạy tuần tự từ trên xuống
 * - Nếu dùng Promise thì không biết lúc nào nó export xong
 */
const fs = require("node:fs");

/* Bước 1: Đọc toàn bộ cây thư mục tại thư mục hiện tại */
// const fileNames = fs.readdirSync("./helpers"); // [ 'index.js', 'joinArray.js', 'truncateTitle.js' ]
// Cần phải loại bỏ chính file index.js vì đây là file chỉ để export
const fileNames = fs
    .readdirSync("./helpers")
    .filter((_fileName) => _fileName !== "index.js"); // [ 'joinArray.js', 'truncateTitle.js' ]

/* Bước 2: Dựa vào [kỳ vọng] Tạo một object có:
    - key: Trùng với tên file (không có `.js`)
    - value: Là giá trị tương ứng được require("...")
 */
const helpers = fileNames.reduce((obj, fileName) => {
    return {
        ...obj,
        [fileName.replace(".js", "")]: require(`./${fileName}`),
    };
}, {});

// Kết quả
// {
//   joinArray: [Function: joinArray],
//   truncateTitle: [Function: truncateTitle]
// }

/* Bước 3: Export object đó */
module.exports = helpers;
```
