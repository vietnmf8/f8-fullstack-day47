````python
// Tạo file: src/middlewares/response.middleware.js

const responseMiddleware = (_, res, next) => {
    // Nếu thành công!
    res.success = (data, status = 200) => {
        res.status(status).json({
            status: "success",
            data,
        });
    };

    // Nếu lỗi
    res.error = (error, status = 400) => {
        res.status(status).json({
            status: "error",
            error,
        });
    };

    next();
};

module.exports = responseMiddleware;


````
