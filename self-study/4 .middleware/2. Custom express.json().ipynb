```python
/* Middleware */
// app.use(express.json());

const customExpress = {
    json() {
        return (req, res, next) => {
            let body = "";
            req.on("data", (chunk) => {
                body += chunk.toString();
            });
            req.on("end", () => {
                req.body = JSON.parse(body);
                next();
            });

            // Vậy tại sao lại có req.body()
            // Express rất hay, nó dùng một object req, res đồng bộ/ truyền xuyên suốt qua tất cả mọi middleware, các middleware đều giống nhau.
            // Vậy chúng ta tạo ra thuộc tính body cho req
            // Sau khi gán thành công rồi next()
        };
    },
};

/* Viết lại middleware express.json() */
app.use(customExpress.json());
```

```python
- Trong thực tế, người ta viết ở trong thư viện Express, tích hợp vào người ta mới viết nhồi nhét thế này và dùng rất gọn
- Nếu chúng ta cố tình bước trước  thì lại thành dở
```

```python
// Tạo src/middlewares/json.middleware.js
const jsonMiddleware = (req, _, next) => {
    let body = "";
    req.on("data", (chunk) => {
        body += chunk.toString();
    });
    req.on("end", () => {
        if (body) {
            req.body = JSON.parse(body);
        }
        next();
    });
};

module.exports = jsonMiddleware;
```

```python
// server.js

/* Sử dụng thực tế */
app.use(jsonMiddleware);
```
