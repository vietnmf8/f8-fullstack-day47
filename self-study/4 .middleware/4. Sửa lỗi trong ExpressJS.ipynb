```python
// Error Handling
// Vấn đề

app.get('/', (req, res) => {
  throw new Error('BROKEN') // Express will catch this on its own.
})

// --> Khi lỗi ứng dụng có thể bị crash, chết
// --> Vậy làm thế nào trong những tình huống xảy ra lỗi ngoại lệ mà chúng ta vẫn giữ được format response của chúng ta

// --> sử dụng tham số thứ tư `err` được viết ở đầu
	`function errorHandler(err, req, res, next)`
// --> giúp chúng ta hứng được lỗi
```

```python
// src/middlewares/errorHandler.middleware.js
const errorHandlerMiddleware = (err, req, res, next) => {
    res.error(err);
};

module.exports = errorHandlerMiddleware;
```

```python
// server.js
app.use("/api", apiRouter);
app.use(errorHandlerMiddleware); // Áp dụng phía sau nơi văng lỗi

// Nguyên lý của middleware có 4 tham số là:
// - Trước đó mà có một thằng throw lỗi là nó sẽ văng sang thằng đằng sau có 4 tham số

// Kết quả

// {
// 	"status": "error",
// 	"error": {}
// }

// Vậy làm sao để người ta kiểm soát được những lỗi này, tình huống này gọi là exception (ngoại lệ)
// - chúng ta không lường trước đó.
// - không may nó rơi vào tính huống đó.
// --> Thường chúng ta xếp loại là lỗi server
// --> trả về lỗi mặc định là 500
```

```python
// Xử lý lỗi 404
// src/middlewares/notFound.middleware.js
const notFoundMiddleware = (req, res) => {
    res.error(
        {
            message: `Cannot ${req.method} ${req.url}`,
        },
        404,
    );
};

module.exports = notFoundMiddleware;
```

```python
//server.js

app.use("/api", apiRouter);
app.use(notFoundMiddleware);
app.use(errorHandlerMiddleware); // Áp dụng phía sau nơi văng lỗi
```
