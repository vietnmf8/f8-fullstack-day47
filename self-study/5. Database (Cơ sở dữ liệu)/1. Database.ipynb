````python
- Database sẽ có thuật toán, cấu trúc dữ liệu để thực hiện việc thực hiện tìm kiếm, lưu trữ, lọc thì có hiệu suất cao hơn
- Database chia thành 2 loại:
	- **RDBMS** (Sử dụng cơ sở dữ liệu quan hệ - `R`elational `D`ata`B`ase `M`anagement `S`ystem)
		+ Sử dụng ngôn ngữ **SQL** để truy vấn
		+ Thể hiện dữ liệu quan hệ thật. VD: con phải là con của dữ liệu cha, thì nó bắt buộc phải có dữ liệu cha thì mới có dữ liệu con (không thể nào tạo được dữ liệu con và dữ liệu cha không tồn tại thật được). Bởi vì nó là quan hệ ràng buộc trong Database
		+ Dữ liệu được thể hiện dưới dạng bảng (Table)
			- Table -> Row/Record (bản ghi) (Hàng) -> Col (Column - Cột)
			- Vì vậy chắc chắn các bản ghi (hàng) đều có số lượng cột như nhau

	- **NoSQL** Không quan hệ (Not Only SQL)
		+ Vì không có tính quan hệ, thì chúng ta có thể giả định quan hệ về mặt logic. 
		+ VD: cha thì phải có trước, rồi mới có con, con là con thuộc cha
		+ Nhưng giả sử cha cha không có trước, hoặc không tồn tại. Thì nó hoàn toàn có thể bị mất đi sự nhất quán của dữ liệu
		+ Dữ liệu được thể hiện dưới dạng  hoặc đôi khi có người gọi là Collection.
		+ Một Row người ta gọi là Document được trình bày dưới dạng (key-value). Thực ra chính là dạng `JSON`
		+ Vì vậy không đảm bảo tính đồng nhất giữa các Document
````

````python
File `db.json` trên đang quản lý `tasks`
- Phần [...] 
  --> gọi là Collection **tasks** (`NoSQL`)
  --> gọi là Table **tasks** (`RDBMS`)

- Phần {...}
  --> gọi là Document (`NoSQL`)
  --> gọi là Row/record (`RDBMS`)

- Đặc điểm:
	+ Document: Mỗi document không cần giống nhau. VD:ư
```json

{
    "tasks": [
        {
            "id": 1,
            "title": "Nấu cơm"
        },
        {
            "id": 2,
			// Có thể thiếu trường ở đây, không cần giống nhau giữa các document
        },
		{
            "id": 3,
            "title": "Nấu cơm",
            "isCompleted": false
			// Tuỳ tiện thêm trường, không bị cấm
        },
    ]
}
```


- Bẫy phỏng vấn: Khi nào bạn dùng NoSQL, khi nào bạn dùng MySQL
- Thường hay trả lời: Thường khi chúng ta cần performance thì dùng MongoDB, cần chuẩn chỉ về quan hệ dữ liệu thì dùng MySQL
- Thật ra: 2 thằng này người ta không đi sử dụng nó vì performance. Mà người ta sử dụng nó bởi sự khác biệt chính: Có quan hệ và Phi quan hệ:
	+ **Có quan hệ**: Rất chặt chẽ về dữ liệu, nhất quán từ đầu đến cuối, đồng nhất tất cả các field các trường nếu thiết lập quan hệ đầy đủ ở tầng Database. Vì nó ở dạng Table
	+ **Phi quan hệ**: Khi không cần tính chặt chẽ, mà cần các document có các field không giống nhau thì sử dụng NoSQL

- Ví dụ chúng ta làm ứng dụng về thương mại điện tử --> ship hàng, hoàn tiền.. --> phức tạp như vậy nên cần tính chặt chẽ về dữ liệu
- Ví dụ có đơn hàng như lại mất sản phẩm
	+ Thằng NoSQL (MongoDB) xoá một sản phẩm đi, các thứ đi thì dữ liệu còn lại vẫn sẽ tồn tại ở đó vì không ràng buộc về quan hệ
	+ Nhưng với thằng quan hệ chặt chẽ: (Nếu muốn xoá dữ liệu quan hệ chặt chẽ với nhau)
		1. Không cho chúng ta xoá
		2. Có thể setup là: Khi xoá thằng cha, thì thằng con bị xoá theo.
		3. Không xảy ra tình huống mồ côi dữ liệu
- Khi mà muốn linh hoạt mà không bị ràng buộc quan hệ thì dùng NoSQL
````
