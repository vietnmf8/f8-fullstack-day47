````python
# ğŸ’Š GIáº¢I PHÃP (Solution)
```javascript
const path = require('node:path');

// âœ… Tá»± Ä‘á»™ng xá»­ lÃ½ Ä‘Ãºng cho má»i há»‡ Ä‘iá»u hÃ nh
const filePath = path.join(__dirname, 'data', 'users.json');
// Windows: C:\projects\myapp\data\users.json
// Mac/Linux: /Users/john/projects/myapp/data/users.json
```

**NÃ³ giÃºp:**
- âœ… Tá»± Ä‘á»™ng dÃ¹ng dáº¥u `/` hoáº·c `\` Ä‘Ãºng vá»›i tá»«ng OS
- âœ… Ná»‘i path an toÃ n, khÃ´ng bá»‹ lá»—i `//` hay `\\`
- âœ… LÃ m viá»‡c vá»›i path dá»… dÃ ng (láº¥y tÃªn file, extension, thÆ° má»¥c cha...)
- âœ… Code cháº¡y Ä‘Æ°á»£c trÃªn Windows, Mac, Linux

---

## ğŸª áº¨N Dá»¤ (Analogy)

### HÃ¬nh áº£nh: **Äá»‹a chá»‰ nhÃ  á»Ÿ Viá»‡t Nam vs Má»¹**
```
KHÃ”NG cÃ³ path module:
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
Báº¡n: "Äi Ä‘áº¿n Ä‘á»‹a chá»‰: Sá»‘ 10, Ä‘Æ°á»ng ABC, quáº­n 1"

Shipper Má»¹: "Háº£? á» Má»¹ pháº£i viáº¿t: 10 ABC Street, District 1"
Shipper Viá»‡t Nam: "á» Viá»‡t Nam pháº£i viáº¿t: Sá»‘ 10, Ä‘Æ°á»ng ABC, quáº­n 1"

â†’ Pháº£i viáº¿t 2 cÃ¡ch khÃ¡c nhau cho 2 nÆ°á»›c!


CÃ“ path module:
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
path.join(): "Äá»ƒ tÃ´i dá»‹ch Ä‘á»‹a chá»‰ cho phÃ¹ há»£p má»—i nÆ°á»›c!"

Báº¡n: path.join('sá»‘ 10', 'Ä‘Æ°á»ng ABC', 'quáº­n 1')

path.join tá»± Ä‘á»™ng dá»‹ch:
- á» Má»¹ â†’ "10/ABC/District1"
- á» Viá»‡t Nam â†’ "10\ABC\Quáº­n1"

â†’ Viáº¿t 1 láº§n, cháº¡y má»i nÆ¡i!
```
````

````python
# ğŸ“ CÃš PHÃP QUAN TRá»ŒNG
1. **`path.join()` - Ná»‘i cÃ¡c pháº§n path**
```javascript
const path = require('node:path');

// âœ… Ná»‘i path an toÃ n
const filePath = path.join('data', 'users', 'profile.json');
// Windows: data\users\profile.json
// Mac/Linux: data/users/profile.json

// âœ… DÃ¹ng vá»›i __dirname (thÆ° má»¥c hiá»‡n táº¡i)
const filePath = path.join(__dirname, 'uploads', 'images', 'avatar.png');
// Windows: C:\projects\myapp\uploads\images\avatar.png
// Mac/Linux: /Users/john/projects/myapp/uploads/images/avatar.png
```

---

**2. `path.resolve()` - Táº¡o absolute path**
```javascript
// âœ… Táº¡o path tuyá»‡t Ä‘á»‘i tá»« thÆ° má»¥c gá»‘c
const absolutePath = path.resolve('data', 'users.json');
// Windows: C:\projects\myapp\data\users.json
// Mac/Linux: /Users/john/projects/myapp/data/users.json

// âœ… DÃ¹ng khi cáº§n path Ä‘áº§y Ä‘á»§
const configPath = path.resolve(__dirname, '..', 'config', 'database.json');
// LÃ¹i 1 cáº¥p (..) rá»“i vÃ o config/database.json
```

---

**3. `path.basename()` - Láº¥y tÃªn file**
```javascript
const filePath = '/users/john/documents/report.pdf';

const fileName = path.basename(filePath);
console.log(fileName); // âœ… report.pdf

const fileNameNoExt = path.basename(filePath, '.pdf');
console.log(fileNameNoExt); // âœ… report
```

---

**4. `path.dirname()` - Láº¥y thÆ° má»¥c cha**
```javascript
const filePath = '/users/john/documents/report.pdf';

const dirPath = path.dirname(filePath);
console.log(dirPath); // âœ… /users/john/documents
```

---

5. **`path.extname()` - Láº¥y extension**
```javascript
const filePath = 'avatar.png';

const ext = path.extname(filePath);
console.log(ext); // âœ… .png
```

---

## ğŸ¯ LUá»’NG HOáº T Äá»˜NG
```
TÃ¬nh huá»‘ng: LÆ°u file upload lÃªn server
â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

1. User upload file "avatar.jpg"
   â”‚
   â–¼
2. Backend cáº§n lÆ°u vÃ o: 
   /uploads/images/user-123/avatar.jpg
   â”‚
   â–¼
3. KHÃ”NG dÃ¹ng path (SAI):
   const uploadPath = __dirname + '/uploads/images/' + userId + '/' + fileName;
   // Windows: Lá»—i! VÃ¬ dÃ¹ng /
   // KhÃ³ Ä‘á»c, dá»… lá»—i
   â”‚
   â–¼
4. DÃ™NG path (ÄÃšNG):
   const uploadPath = path.join(__dirname, 'uploads', 'images', userId, fileName);
   // âœ… Windows: C:\projects\myapp\uploads\images\user-123\avatar.jpg
   // âœ… Mac/Linux: /projects/myapp/uploads/images/user-123/avatar.jpg
```
````
